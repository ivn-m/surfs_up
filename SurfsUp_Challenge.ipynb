```python
# Dependencies
import numpy as np
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
for m in Measurement.__table__.columns:
    print(m)
```

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june_temps = session.query(Measurement).filter(extract('month',Measurement.date) == 6).all()

#june_temps
```

```python
#  3. Convert the June temperatures to a list.
june_temps_list = [temp.tobs for temp in june_temps]

june_temps_list
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_temps_df = pd.DataFrame(june_temps_list, columns= ['June Temperatures'])

#june_temps_df
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_temps_summary_stats = june_temps_df.describe()

june_temps_summary_stats
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
december_temps = session.query(Measurement).filter(extract('month',Measurement.date) == 12).all()

december_temps
```

```python
# 7. Convert the December temperatures to a list.
december_temps_list = [temp.tobs for temp in december_temps]

december_temps_list
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
december_temps_df = pd.DataFrame(december_temps_list, columns= ['December Temperatures'])

#december_temps_df
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
december_temps_summary_stats = december_temps_df.describe()

december_temps_summary_stats
```

```python
# June precipitation
june_prcp = session.query(Measurement.date, Measurement.prcp).filter(extract('month',Measurement.date) == 6).all()
june_df = pd.DataFrame(june_prcp, columns=['date','June precipitation'])
june_df.set_index(june_df['date'], inplace=False)
june_df= june_df.sort_index()

june_df.describe()
```

```python
# December precipitation
december_prcp = session.query(Measurement.date, Measurement.prcp).filter(extract('month',Measurement.date) == 12).all()
december_df = pd.DataFrame(december_prcp, columns=['date','December precipitation'])
december_df.set_index(december_df['date'], inplace=False)
december_df= december_df.sort_index()

december_df.describe()
```
